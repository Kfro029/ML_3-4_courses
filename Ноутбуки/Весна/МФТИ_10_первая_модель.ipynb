## Возьмем несколько признаков

```python
features = [
    'store',
    'dayofweek',
    'customers',
    'open',
    'promo',
]

target = ['sales']
```

### 🧠 Упражнение: обучение модель KNN (ответ)



1. Инициализируйте модель KNN для задачи регрессии


```python
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor()
model
```

2. Обучите модель на тренировочных данных


```python
model.fit(
    X=df_train[features],
    y=df_train[target]
)
```

3. Сделайте предсказаний для тестовых объектов

```python
pred = model.predict(df_test[features])
```

4. Соедините предсказания и истинные значения в таблицу


```python
results = pd.DataFrame(pred, columns=['pred'])
results.head()
```

```python
results['target'] = df_test['sales'].reset_index(drop=True)
results.head()
```

5. Посчитайте метрику RMSE

```python
from sklearn.metrics import mean_squared_error

print(np.sqrt(mean_squared_error(results['target'], results['pred'])))  # RMSE
print(mean_squared_error(results['target'], results['pred'], squared=False))  # RMSE
```
